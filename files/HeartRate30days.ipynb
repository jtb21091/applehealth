```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'Heart Rate.csv'
heart_rate_data = pd.read_csv(file_path)

# Convert Date/Time to datetime and sort
heart_rate_data['Date/Time'] = pd.to_datetime(heart_rate_data['Date/Time'])
heart_rate_data = heart_rate_data.sort_values('Date/Time')

# Calculate overall average and standard deviation
overall_avg = heart_rate_data['Avg (count/min)'].mean()
std_dev = heart_rate_data['Avg (count/min)'].std()

# Display the average and standard deviation
print(f"Overall Average Heart Rate: {overall_avg:.2f} count/min")
print(f"Standard Deviation: {std_dev:.2f} count/min")

# Plot the chart
plt.figure(figsize=(12, 6))
plt.plot(heart_rate_data['Date/Time'], heart_rate_data['Avg (count/min)'], label='Heart Rate Data', linewidth=2)
plt.axhline(y=overall_avg, color='green', linestyle='-', label='Average Heart Rate')
plt.axhline(y=overall_avg + 2 * std_dev, color='red', linestyle='--', label='2 Std Above')
plt.axhline(y=overall_avg - 2 * std_dev, color='blue', linestyle='--', label='2 Std Below')

# Adding labels, title, and legend
plt.xlabel('Date/Time')
plt.ylabel('Heart Rate (count/min)')
plt.title('Heart Rate Data: Average and Standard Deviation Thresholds')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
