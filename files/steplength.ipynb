```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = 'steplength.xlsx'
df = pd.read_excel(file_path)

# Using the correct column name 'step length'
step_length = df['Step length']

# Calculate statistics
mean_step_length = step_length.mean()
std_step_length = step_length.std()
minus_2std = mean_step_length - 2 * std_step_length
plus_2std = mean_step_length + 2 * std_step_length

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df.index, step_length, label='Step Length', alpha=0.6)

# Plot average, -2std, and +2std lines
plt.axhline(mean_step_length, color='r', linestyle='-', label='Average')
plt.axhline(minus_2std, color='g', linestyle='-', label='-2 STD')
plt.axhline(plus_2std, color='b', linestyle='-', label='+2 STD')

# Add labels and legend
plt.xlabel('Index')
plt.ylabel('Step Length')
plt.title('Step Length Scatter Plot with Mean and 2 STD Lines')
plt.legend()

# Show the plot
plt.show()
```
